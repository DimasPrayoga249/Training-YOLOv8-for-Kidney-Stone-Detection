Hyperparameter -> Nilai yang ditentukan sebelum training untuk mengatur bagaimana proses pembelajaran berlangsung

1. Seed(Hyperparameter) = Digunakan untuk mengatur random initialization agar hasil eksperimen dapat direprodusi.

2. Epochs(Hyperparameter) = Menentukan berapa kali seluruh dataset dilatih pada model.

3. Optimizer(Hyperparameter) = Menentukan algoritma optimasi yang digunakan untuk memperbarui weight.

4. Learning_Rate(Hyperparameter) = Learning rate awal, mengatur besar langkah saat update weight.

```python
from logging import warning # To display warning messages
import pandas as pd # Reading datasets
import numpy as np # Mathematical computations
import PIL # Image Processing and Manipulation
from PIL import Image
from IPython.display import display # Display objects clearly in Jupyter Notebook
import matplotlib.pyplot as plt # Creating visualizations and plots.
from glob import glob # Find files in directories using pattern matching.
import random # Generates random values.
import cv2 # Widely used for computer vision and image processing.
import warnings # Controls how Python displays warning messages.
warnings.simplefilter('ignore') #
```

```python
# Pick image randomly from dataset
root_path = f'{path}/test/images/*'
num_samples = 10
images_data = glob(root_path)
random_image = random.sample(images_data, num_samples)
plt.figure(figsize=(12,24))

# Show image with grid shape using matplotlib
for i in range(num_samples):
  plt.subplot(5,2,i+1) # Changed subplot grid to 2x5 to accommodate 10 images
  plt.imshow(cv2.imread(random_image[i]))
  plt.axis('off')
```

```python
# Pick image from dataset
images = []

# Process every image that pick randomly
for i in range(num_samples):
  yolo_outputs = model.predict(random_image[i])
  output = yolo_outputs[0]
  box = output.boxes
  names = output.names
  print("***********")

  # Process every object that detected with one image
  for j in range(len(box)):
    labels = names[box.cls[j].item()]
    coordinates = box.xyxy[j].tolist()
    confidence = np.round(box.conf[j].item(), 2)
    print(f'In this image {len(box)} stone has been detected.')
    print(f'Stone {j+1} is: {labels}')
    print(f'Cooordinates: {coordinates}')
    print(f'Confidence is: {confidence}')
    print("***********")

  # Move this line outside the inner loop to append only one image per original sample
  images.append(output.plot()[:, :, ::-1])
```

```python
# Displays multiple images while detecting kidney stone objects.
plt.figure(figsize=(12,24))
for i, img in enumerate(images) :
  plt.subplot(5,2,i+1)
  plt.imshow(img)
  plt.axis('off')
```
